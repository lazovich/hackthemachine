```python
%matplotlib nbagg
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
data = {}
path = "/home/ec2-user/clean_data/"
prefix = "classC_ship2"

for f in os.listdir(path):
    if f.startswith(prefix) and f.endswith(".csv"):
        print("Parsing file: " + f)
        data[f] = pd.read_csv(path + f)
```

```python
df = data['classC_ship2_SSDG_sort.csv']
df = df.drop_duplicates().drop('Unnamed: 0', axis=1)

times = pd.to_datetime(df['DateTime'])
df['timestamp'] = times
df = df.set_index('timestamp')
```

```python
#df = df.groupby(['timestamp']).first().reset_index().set_index('timestamp')
df = df.sort_index()
df
```

```python
start = 0#'2017-03-01'
end = -1#'2017-03-20'
```

```python
%matplotlib inline
for col in df.columns[10:20]:
    print(col)
    try: 
        plt.figure()
        df[col][start:end].plot()
    except:
        pass
```

```python
from sklearn.preprocessing import Imputer
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)

numerical_data = df.drop('Indicator', axis=1).drop('DateTime', axis=1).drop('timeFormat', axis=1)[start:end]
X = imp.fit_transform(np.array(numerical_data, 'float'))

from sklearn.decomposition import SparsePCA as PCA

pca = PCA(n_components=4)
pca.fit(X)

X_pca = pca.transform(X).transpose()
```

```python
print(X.transpose().shape)
print(X_pca.shape)
```

```python
line_data = np.vstack([X.transpose(), X_pca[0]])
new_data = []
for row in range(line_data.shape[0]):
    new_data.append(
        np.clip(line_data[row, :] / np.max(line_data[row, :]), 0, 1)
    )
new_data = np.array(new_data)
print(new_data.shape)
print(line_data.shape)
import datetime
from dateutil import parser
dt = parser.parse("Aug 28 1999 12:00AM")
times = [parser.parse(a) for a in df.timeFormat[:-1]]
```

```python
plt.figure(figsize=(15,5))
lines = plt.plot(times, new_data.transpose())
for l in lines:
    plt.setp(l, linewidth=0.8)
plt.setp(lines[-3], linewidth=5, color="black")
plt.ylim(0.2, 1.1)
p = 50070
plt.xlim(times[p], times[p+500])
plt.ylabel("Normalized Sensor Data")
plt.xlabel("Time")
plt.gcf().autofmt_xdate()
plt.yticks([])
#plt.plot(line_data.transpose())
```
