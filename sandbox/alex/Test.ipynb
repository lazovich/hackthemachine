```python
data = {}
path = "/home/ec2-user/hackathon_data/"
prefix = "classC_ship2"

for f in os.listdir(path):
    if f.startswith(prefix) and f.endswith(".csv"):
        print("Parsing file: " + f)
        data[f] = pd.read_csv(path + f)
```

```python
columns = {}
for fname in data:
    columns[fname] = data[fname].columns.values
columns
```

```python
prefixSplit = {}
for name in columns['classC_ship2_MPDE.csv']:
    prefix = name.split(" ")[0]
    print(prefix)
    if prefix not in prefixSplit:
        prefixSplit[prefix] = []
    prefixSplit[prefix].append(name.split(" ")[1:])
```
