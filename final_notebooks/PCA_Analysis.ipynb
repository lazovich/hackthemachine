```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# Collect data files into Pandas DataFrames

data = {}
path = "/home/ec2-user/clean_data/"
prefix = ""

for f in os.listdir(path):
    if f.startswith(prefix) and f.endswith(".csv"):
        print("Parsing file: " + f)
        data[f] = pd.read_csv(path + f)
```

```python
# A few steps to clean the data and sort it into a time indexed table

df = data['classC_ship2_MRG_sort.csv']
df = df.drop_duplicates()
df =df.drop('Unnamed: 0', axis=1)

times = pd.to_datetime(df['DateTime'])
df['timestamp'] = times
df = df.set_index('timestamp')
```
